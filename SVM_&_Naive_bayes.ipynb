#**SVM & Naive bayes**

**1. What is a Support Vector Machine (SVM)?**
- A **Support Vector Machine (SVM)** is a powerful machine learning algorithm used for classification and sometimes regression. It works by finding the best boundary (called a **hyperplane**) that separates data points of different classes. The goal is to choose the boundary that leaves the biggest gap, or **margin**, between the two classes. The data points that are closest to this boundary are called **support vectors**, and they are the most important in defining the line. SVM can also handle more complex data by using something called **kernels**, which allows it to work well even when the data isn’t linearly separable.


**2. What is the difference between Hard Margin and Soft Margin SVM?**
- The main difference between Hard Margin and Soft Margin SVM is how strict the model is about separating the classes. Hard Margin SVM tries to find a perfect boundary that separates the classes without any mistakes, meaning no points are allowed on the wrong side of the line. This only works when the data is perfectly separable. Soft Margin SVM, on the other hand, allows some misclassifications or overlap between classes so it can handle noisy or overlapping data better. It balances finding a wide margin with keeping classification errors low, making it more practical for real-world problems.

**3. What is the mathematical intuition behind SVM?**
- The mathematical intuition behind **Support Vector Machine (SVM)** is to find the line (or hyperplane) that best separates the classes by creating the **largest possible margin** between them. This margin is the distance between the hyperplane and the closest data points from each class, called **support vectors**. SVM tries to **maximize this margin** while minimizing classification errors. When data isn’t linearly separable, SVM uses a technique called the **kernel trick** to transform the data into a higher-dimensional space where a linear separator can be found. The goal is to solve an optimization problem that balances keeping the margin wide and allowing a few errors (if using a soft margin).


**4. What is the role of Lagrange Multipliers in SVM?**
- The role of **Lagrange Multipliers** in SVM is to help solve the **optimization problem** that finds the best separating hyperplane. Instead of directly solving the problem with constraints (like keeping all data points on the correct side of the margin), SVM uses Lagrange Multipliers to convert it into a **dual problem** that is easier to handle mathematically. These multipliers determine the importance of each data point in defining the margin — and only a few points (called **support vectors**) have non-zero multipliers. This approach allows SVM to efficiently handle constraints and also makes it possible to use **kernel functions** in the dual form to deal with non-linear data.


**5. What are Support Vectors in SVM?**
- **Support Vectors** in SVM are the **data points that lie closest to the decision boundary (hyperplane)**. They are the most important points in the dataset because they directly affect the position and orientation of the hyperplane. The SVM algorithm uses only these points to determine the optimal boundary that separates the classes with the **maximum margin**. If we remove a support vector, the decision boundary may change, but removing other (non-support) points usually won’t affect it. So, support vectors are the key elements that “support” or define the structure of the SVM model.


**6. What is a Support Vector Classifier (SVC)?**
- A **Support Vector Classifier (SVC)** is the classification version of the **Support Vector Machine (SVM)** algorithm. It is used to separate data into different **classes** by finding the best boundary (called a hyperplane) that maximizes the margin between classes. SVC can handle both **linearly separable** and **non-linearly separable** data. For non-linear problems, it uses something called the **kernel trick** to transform the data into a higher-dimensional space where a linear boundary can be found. In simple terms, SVC is a powerful and flexible tool for making accurate predictions in classification tasks.


**7. What is a Support Vector Regressor (SVR)?**
- A **Support Vector Regressor (SVR)** is the regression version of Support Vector Machine (SVM), used to predict **continuous values** instead of class labels. Unlike regular regression models that try to minimize the error for every point, SVR tries to fit the best line (or curve) within a certain **margin of tolerance** around the true values. This margin is called the **epsilon-tube**, and the model focuses only on the data points that fall **outside** this tube, treating them as important (like support vectors in classification). SVR is useful when we want a model that is robust to small errors and focuses more on capturing the overall trend rather than perfectly fitting every point.


**8. What is the Kernel Trick in SVM?**
- The **Kernel Trick** in SVM is a clever method that allows the algorithm to solve problems where the data is **not linearly separable** by transforming it into a **higher-dimensional space** without actually computing the transformation directly. Instead of mapping the data to this new space manually, the kernel trick uses a **mathematical function (kernel)** to calculate the **dot product** of the data in that higher-dimensional space. This makes it possible for SVM to find a linear separator in the transformed space, which corresponds to a **nonlinear boundary** in the original space. Common kernels include the **polynomial**, **RBF (Gaussian)**, and **sigmoid** kernels.


**9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel?**
- The **Linear Kernel** is the simplest and fastest option, used when data can be separated with a straight line, making it ideal for linearly separable problems or high-dimensional data like text. The **Polynomial Kernel** allows the model to create curved decision boundaries by raising the input features to a certain power, making it suitable for problems with more complex patterns, but it may overfit if the degree is too high. The **RBF (Radial Basis Function) Kernel** is the most flexible, capable of modeling very complex and nonlinear relationships by considering the distance between points, though it requires careful tuning of parameters. In summary, Linear is fast and simple, Polynomial captures moderate complexity, and RBF handles highly nonlinear data.


**10. What is the effect of the C parameter in SVM?**
- The **C parameter** in SVM controls the **trade-off between achieving a low error on the training data and having a simple, smooth decision boundary**. A **small C** value makes the SVM allow more misclassifications (soft margin), which means it tries to create a wider margin and generalizes better but might misclassify some training points. A **large C** value tries to classify all training points correctly by giving more importance to minimizing errors, which can lead to a narrower margin and potentially overfitting. So, **C balances flexibility and accuracy** — smaller values encourage simplicity and better generalization, while larger values fit the training data more tightly.


**11. What is the role of the Gamma parameter in RBF Kernel SVM?**
- The **Gamma** parameter in an RBF Kernel SVM controls how much influence a single training example has. It defines the **reach or spread** of a single data point’s influence on the decision boundary. A **small gamma** means each point’s influence is far-reaching, making the decision boundary smoother and less complex. A **large gamma** means each point’s influence is very close, which can create a more complex and wiggly boundary that fits the training data tightly but may overfit. In short, gamma controls the shape of the decision boundary: low gamma = smooth and simple, high gamma = detailed and complex.


**12. What is the Naïve Bayes classifier, and why is it called "Naïve"?**
- The **Naïve Bayes classifier** is a simple and fast probabilistic machine learning algorithm used for classification. It’s based on **Bayes’ Theorem**, which calculates the probability of a class given the features. The “naïve” part comes from the assumption that all features are **independent of each other** — which is usually not true in real life but makes the calculations much simpler and faster. Despite this strong and often unrealistic assumption, Naïve Bayes works surprisingly well for many problems like spam detection and text classification.

**14. Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes?**
- The three main types of Naïve Bayes differ based on the kind of data they work best with. **Gaussian Naïve Bayes** assumes the features follow a normal (bell-curve) distribution, so it’s great for continuous data like height or weight. **Multinomial Naïve Bayes** works well with **count data**—for example, how many times a word appears in a document—making it popular for text classification with word frequencies. **Bernoulli Naïve Bayes** is used for **binary data**, where features are just present or absent (like whether a word appears or not in a text). So, each type fits different kinds of data based on their mathematical assumptions.


**15. When should you use Gaussian Naïve Bayes over other variants?**
- We should use **Gaussian Naïve Bayes** when our features are **continuous numerical values** that roughly follow a normal (bell-shaped) distribution, like measurements of height, weight, or temperature. It’s the best choice if our data isn’t just counts or binary, and we want the model to assume each feature’s values spread out in a way that fits a Gaussian curve. If our data is categorical or consists of counts, other variants like Multinomial or Bernoulli Naïve Bayes would be more suitable.


**16. What are the key assumptions made by Naïve Bayes?**
- Naïve Bayes makes a few simple but important assumptions. First, it assumes that all the features are **independent of each other**, meaning one feature doesn’t affect or depend on another. Second, it assumes that each feature contributes equally and separately to deciding the class. Lastly, depending on the type of data, it assumes the features follow a certain kind of distribution (like normal distribution for continuous data). These assumptions make the math easier, but they might not always perfectly match real-world data.


**17. What are the advantages and disadvantages of Naïve Bayes?**
- Naïve Bayes is simple, fast, and works well with large and high-dimensional data, like text. It also needs less training data and often performs well even if its assumptions aren’t perfect. However, it assumes all features are independent, which isn’t usually true and can reduce accuracy. It can also struggle if features are closely related or if the data doesn’t fit its assumed distribution. Overall, it’s a great starting model but might not always be the most accurate.


**18. Why is Naïve Bayes a good choice for text classification?**
- Naïve Bayes is a good choice for text classification because it works well with **high-dimensional data** like lots of words, and it’s very fast to train even on large text collections. It handles the way words appear in documents using simple probabilities and doesn’t need a lot of data to learn. Plus, even though it assumes words are independent (which isn’t always true), it still gives surprisingly accurate results for tasks like spam detection or sentiment analysis.


**19. Compare SVM and Naïve Bayes for classification tasks?**
- SVM and Naïve Bayes are both popular classifiers but work differently. **SVM** tries to find the best boundary that separates classes by maximizing the margin and works well for complex, high-dimensional data, especially when classes aren’t easily separated. It can handle nonlinear data using kernels but can be slower to train. **Naïve Bayes**, on the other hand, is simpler and faster, based on probability and the assumption that features are independent. It works great for text and high-dimensional data but may be less accurate if features are strongly related. So, SVM is powerful for complicated boundaries, while Naïve Bayes is fast and effective for large, simple problems like text classification.


**20. How does Laplace Smoothing help in Naïve Bayes?**
- **Laplace Smoothing** helps Naïve Bayes by fixing the problem of **zero probabilities**. Sometimes, a feature or word might not appear in the training data for a certain class, which would make its probability zero and cause the whole prediction to fail (because multiplying by zero gives zero). Laplace Smoothing adds a small constant (usually 1) to all feature counts, ensuring no probability is exactly zero. This makes the model more robust, especially when dealing with words or features that didn’t appear in the training examples.

**21. Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy.**

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X = iris.data
y = iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_clf = SVC()

svm_clf.fit(X_train, y_train)

y_pred = svm_clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of SVM classifier on Iris dataset: {accuracy:.2f}")
```

**22.Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then compare their accuracies.**

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

wine = datasets.load_wine()
X = wine.data
y = wine.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_linear = SVC(kernel='linear')
svm_linear.fit(X_train, y_train)
y_pred_linear = svm_linear.predict(X_test)
accuracy_linear = accuracy_score(y_test, y_pred_linear)

svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(X_train, y_train)
y_pred_rbf = svm_rbf.predict(X_test)
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)

print(f"Accuracy with Linear kernel: {accuracy_linear:.2f}")
print(f"Accuracy with RBF kernel: {accuracy_rbf:.2f}")
```

**23.Write a Python program to train an SVM Regressor (SVR) on a housing dataset and evaluate it using Mean Squared Error (MSE).**

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler

housing = fetch_california_housing()
X = housing.data
y = housing.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).ravel()

svr = SVR(kernel='rbf')
svr.fit(X_train_scaled, y_train_scaled)

y_pred_scaled = svr.predict(X_test_scaled)

y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1))
y_test_original = scaler_y.inverse_transform(y_test.reshape(-1, 1))

mse = mean_squared_error(y_test_original, y_pred)
print(f"Mean Squared Error (MSE) of SVR on California housing dataset: {mse:.2f}")
```

**24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

X = X[y != 2]
y = y[y != 2]

svm_poly = SVC(kernel='poly', degree=3, coef0=1, C=1)
svm_poly.fit(X, y)

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

Z = svm_poly.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolors='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('SVM with Polynomial Kernel Decision Boundary')
plt.show()
```

**27.Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

X = X[y != 2]
y = y[y != 2]

C_values = [0.1, 1, 10, 100]

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

plt.figure(figsize=(12, 10))

for i, C in enumerate(C_values, 1):

    svm_clf = SVC(kernel='linear', C=C)
    svm_clf.fit(X, y)

    Z = svm_clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)


    plt.subplot(2, 2, i)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.title(f"SVM with C={C}")
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()
```

**29.Write a Python program to apply feature scaling before training an SVM model and compare results with unscaled data.**

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
svm_unscaled = SVC()
svm_unscaled.fit(X_train, y_train)
y_pred_unscaled = svm_unscaled.predict(X_test)
accuracy_unscaled = accuracy_score(y_test, y_pred_unscaled)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm_scaled = SVC()
svm_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = svm_scaled.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)

print(f"Accuracy without scaling: {accuracy_unscaled:.2f}")
print(f"Accuracy with scaling: {accuracy_scaled:.2f}")
```

**30. Write a Python program to train a Gaussian Naïve Bayes model and compare the predictions before and after Laplace Smoothing.**

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

gnb_no_smoothing = GaussianNB(var_smoothing=1e-9)
y_pred_no_smoothing = gnb_no_smoothing.predict(X_test)

gnb_with_smoothing = GaussianNB(var_smoothing=1e-2)
gnb_with_smoothing.fit(X_train, y_train)
y_pred_with_smoothing = gnb_with_smoothing.predict(X_test)

accuracy_no_smoothing = accuracy_score(y_test, y_pred_no_smoothing)
accuracy_with_smoothing = accuracy_score(y_test, y_pred_with_smoothing)

print("Accuracy without Laplace Smoothing:", accuracy_no_smoothing)
print("Accuracy with Laplace Smoothing:", accuracy_with_smoothing)
```

**31. Write a Python program to train an SVM Classifier and use GridSearchCV to tune the hyperparameters (C, gamma, kernel).**

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm = SVC()

param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 'auto', 0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly']
}

grid_search = GridSearchCV(estimator=svm, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

grid_search.fit(X_train, y_train)

print("Best parameters found:", grid_search.best_params_)

y_pred = grid_search.best_estimator_.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy with best parameters: {accuracy:.2f}")
```

**32.Write a Python program to train an SVM Classifier on an imbalanced dataset and apply class weighting and check it improve accuracy.**

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score

X, y = make_classification(n_samples=1000, n_features=20, n_classes=2,
                           weights=[0.9, 0.1], flip_y=0, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm_no_weight = SVC()
svm_no_weight.fit(X_train, y_train)
y_pred_no_weight = svm_no_weight.predict(X_test)
acc_no_weight = accuracy_score(y_test, y_pred_no_weight)

svm_weighted = SVC(class_weight='balanced')
svm_weighted.fit(X_train, y_train)
y_pred_weighted = svm_weighted.predict(X_test)
acc_weighted = accuracy_score(y_test, y_pred_weighted)

print(f"Accuracy without class weighting: {acc_no_weight:.2f}")
print(f"Accuracy with class weighting: {acc_weighted:.2f}")

print("\nClassification report WITHOUT class weighting:")
print(classification_report(y_test, y_pred_no_weight))

print("\nClassification report WITH class weighting:")
print(classification_report(y_test, y_pred_weighted))
```

**33. Write a Python program to implement a Naïve Bayes classifier for spam detection using email data.**

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

data = {
    'email': [
        "Win a free iPhone now!",
        "Meeting at 3 PM tomorrow",
        "Congratulations, you've won a lottery!",
        "Don't forget to submit the report",
        "Claim your free gift card today!",
        "Let's catch up over coffee"
    ],
    'label': ['spam', 'ham', 'spam', 'ham', 'spam', 'ham']
}

df = pd.DataFrame(data)

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['email'])
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = MultinomialNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

new_email = ["You have won a free vacation!"]
new_email_transformed = vectorizer.transform(new_email)
prediction = model.predict(new_email_transformed)
print("\nPrediction for new email:", prediction[0])
```

**40. Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and compare accuracy.**

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm = SVC(kernel='linear')
svm.fit(X_train, y_train)
y_pred_full = svm.predict(X_test)
accuracy_full = accuracy_score(y_test, y_pred_full)

rfe = RFE(estimator=SVC(kernel='linear'), n_features_to_select=2)
rfe.fit(X_train, y_train)

X_train_rfe = rfe.transform(X_train)
X_test_rfe = rfe.transform(X_test)

svm_rfe = SVC(kernel='linear')
svm_rfe.fit(X_train_rfe, y_train)
y_pred_rfe = svm_rfe.predict(X_test_rfe)
accuracy_rfe = accuracy_score(y_test, y_pred_rfe)

print(f"Accuracy without feature selection: {accuracy_full:.2f}")
print(f"Accuracy with RFE (top 2 features): {accuracy_rfe:.2f}")
```

**44.Write a Python program to train an SVM Regressor (SVR) and evaluate its performance using Mean Absolute Error (MAE) instead of MSE.**

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import StandardScaler

housing = fetch_california_housing()
X = housing.data
y = housing.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).ravel()

svr = SVR(kernel='rbf')
svr.fit(X_train_scaled, y_train_scaled)

y_pred_scaled = svr.predict(X_test_scaled)

y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1))
y_test_original = scaler_y.inverse_transform(y_test.reshape(-1, 1))

mae = mean_absolute_error(y_test_original, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
```

**45.Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC score.**

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

data = datasets.load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

gnb = GaussianNB()
gnb.fit(X_train, y_train)

y_prob = gnb.predict_proba(X_test)[:, 1]

roc_auc = roc_auc_score(y_test, y_prob)
print(f"ROC-AUC Score: {roc_auc:.4f}")

fpr, tpr, thresholds = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, label=f'ROC curve (area = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Gaussian Naive Bayes')
plt.legend(loc='lower right')
plt.show()
```

**46. Write a Python program to train an SVM Classifier and visualize the Precision-Recall Curve.**

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import precision_recall_curve, average_precision_score

data = datasets.load_breast_cancer()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

svm = SVC(kernel='linear', probability=True, random_state=42)
svm.fit(X_train, y_train)

y_scores = svm.predict_proba(X_test)[:, 1]

precision, recall, thresholds = precision_recall_curve(y_test, y_scores)

avg_precision = average_precision_score(y_test, y_scores)

plt.figure(figsize=(8,6))
plt.plot(recall, precision, label=f'Average Precision = {avg_precision:.2f}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve - SVM Classifier')
plt.legend(loc='best')
plt.grid()
plt.show()
```
